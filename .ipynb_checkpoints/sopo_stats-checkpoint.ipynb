```python
import numpy as np


# --- Data Cleaning Functions ---

# Function to clean count columns (handle -1 and commas)
def clean_count_column(series):
    # Replace the string '-1' with NaN (as it is used for suppressed data)
    # Also handle the string '-1.0' which appears in some columns
    series = series.astype(str).str.replace(r'^-1$', 'NaN', regex=True).str.replace(r'^-1\.0$', 'NaN', regex=True)
    # Remove commas from thousands separators
    series = series.str.replace(',', '', regex=False)
    # Convert to numeric, coercing any remaining non-numeric (like 'nan' strings or other placeholders) to NaN
    return pd.to_numeric(series, errors='coerce')

# Function to clean percentage columns (handle -1.0%, remove %, convert to decimal)
def clean_pct_column(series):
    # Replace the string '-1.0%' or '-1.0' or just '-1' with NaN
    series = series.astype(str).str.replace(r'^-1\.0%', 'NaN', regex=True).str.replace(r'^-1\.0$', 'NaN', regex=True).str.replace(r'^-1$', 'NaN', regex=True)
    # Remove the '%' sign
    series = series.str.replace('%', '', regex=False)
    # Convert to numeric, coercing any remaining non-numeric to NaN
    numeric_series = pd.to_numeric(series, errors='coerce')
    # Convert percentage to decimal (e.g., 90.0 -> 0.900)
    return numeric_series / 100.0

# --- Apply Cleaning ---

# Columns to clean as Counts
count_cols = [
    'students_required_to_test_science', 'students_tested_science', 
    'students_below_expectations_science', 'students_at_above_expectations_science', 
    'students_required_to_test_english', 'students_tested_english', 
    'students_below_expectations_english', 'students_at_above_expectations_english',
    'students_required_to_test_ela', 'students_tested_ela', 
    'students_below_expectations_ela', 'students_at_above_expectations_ela', 
    'students_required_to_test_math', 'students_tested_math', 
    'students_below_expectations_math', 'students_at_above_expectations_math',
    'students_post_secondary', 'population_group_enrollment', 'total_students'
]
sp_df[count_cols] = sp_df[count_cols].apply(clean_count_column)

# Columns to clean as Percentages (to decimal)
pct_cols = [
    'pct_students_tested_science', 'pct_below_expectations_science', 'pct_at_above_expectations_science', 
    'pct_students_tested_english', 'pct_below_expectations_english', 'pct_at_above_expectations_english', 
    'pct_students_tested_ela', 'pct_below_expectations_ela', 'pct_at_above_expectations_ela', 
    'pct_students_tested_math', 'pct_below_expectations_math', 'pct_at_above_expectations_math', 
    'pct_students_post_secondary'
]
sp_df[pct_cols] = sp_df[pct_cols].apply(clean_pct_column)

# --- Select and Format Final Statistics ---

# 1. Multi-Year Trends for Proficiency (At or Above Expectations) and Enrollment
proficiency_cols = [
    'year', 
    'total_students', 
    'pct_at_above_expectations_science', 
    'pct_at_above_expectations_ela', 
    'pct_at_above_expectations_math'
]
trend_df = sp_df[proficiency_cols].set_index('year').sort_index()

# Rename and format the trend columns for the final table (to percentage string)
trend_df = trend_df.rename(columns={
    'total_students': 'Total Students',
    'pct_at_above_expectations_science': 'Science',
    'pct_at_above_expectations_ela': 'ELA (Reading/Writing)',
    'pct_at_above_expectations_math': 'Mathematics'
})

# Format percentages to string with one decimal place and a '%' sign
for col in ['Science', 'ELA (Reading/Writing)', 'Mathematics']:
    trend_df[col] = trend_df[col].apply(lambda x: f"{x * 100:.1f}%" if pd.notna(x) else 'N/A')

# Format total students with comma separators
trend_df['Total Students'] = trend_df['Total Students'].apply(lambda x: f"{int(x):,}" if pd.notna(x) else 'N/A')

# 2. Key Statistics Snapshot (Most Recent Year with Data - 2024/2025)
# Use 2024 for Enrollment and Post-Secondary data as 2025 is missing
# Use 2025 for Test Data as it is the most recent
recent_year = 2025
snapshot_df_2025 = sp_df[sp_df['year'] == recent_year].iloc[0]

# --- Extract individual metrics for a summary table ---
metrics_summary = {
    'Metric': [
        'Total Students (2024)', 
        'Post-Secondary Enrollment (2024)',
        'Science Test Participation (2025)',
        'ELA Test Participation (2025)',
        'Mathematics Test Participation (2025)'
    ],
    'Value': [
        # Total Students (using 2024 due to 2025 NaN)
        f"{sp_df[sp_df['year'] == 2024]['total_students'].iloc[0]:,.0f}",
        # Post-Secondary Enrollment (using 2024 due to 2025 NaN)
        f"{sp_df[sp_df['year'] == 2024]['pct_students_post_secondary'].iloc[0] * 100:.1f}%",
        # Test Participation (using 2025)
        f"{snapshot_df_2025['pct_students_tested_science'] * 100:.1f}%",
        f"{snapshot_df_2025['pct_students_tested_ela'] * 100:.1f}%",
        f"{snapshot_df_2025['pct_students_tested_math'] * 100:.1f}%"
    ]
}
snapshot_df = pd.DataFrame(metrics_summary)

# Save the final dataframes to CSV
trend_df.to_csv('south_portland_proficiency_trend.csv')
snapshot_df.to_csv('south_portland_key_metrics_snapshot.csv', index=False)

print("Proficiency Trend Data:")
print(trend_df.to_markdown(numalign="left", stralign="left"))
print("\nKey Metrics Snapshot Data:")
print(snapshot_df.to_markdown(index=False, numalign="left", stralign="left"))
```

### What is ESSA?
ESSA, a federal education law, creates a long-term, stable commitment to equal opportunity for all students. The new ESSA law provides states additional flexibility and encourages states and schools to innovate and build upon past success while at the same time, being accountable for students. This increased flexibility gives states and school districts an opportunity to tailor strategies and interventions to meet the needs of students and schools.

Assessments are performed in 5th grade, 8th grade, and in high school. 

## ESSA Performance: District Wide

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the CSV file
df = pd.read_csv('essa_data/2025_maine_essa.csv')

# Define the latest year and key performance columns
latest_year = df['year'].max()
perf_cols = [
    'pct_at_above_expectations_science',
    'pct_at_above_expectations_ela',
    'pct_at_above_expectations_math'
]

# Filter for the latest year, All Schools, All Students, All Grades
district_aggregate_df = df[
    (df['year'] == latest_year) &
    (df['school'] == 'All Schools') &
    (df['students'] == 'All Students') &
    (df['grades'] == 'All Grades')
].copy()

# Drop rows where any of the percentage columns is NaN
district_aggregate_df.dropna(subset=perf_cols, inplace=True)

# 1. Calculate Statewide Performance
state_average = district_aggregate_df[perf_cols].mean().to_dict()
state_average['district'] = 'State Average'

state_median = df[perf_cols].median().to_dict()
state_median['district'] = 'State Median'

# 2. Extract South Portland Schools performance
sp_df = district_aggregate_df[district_aggregate_df['district'] == 'South Portland Public Schools']
sp_performance = sp_df[perf_cols].iloc[0].to_dict()
sp_performance['district'] = 'South Portland Schools'

# 3. Combine the data
comparison_data = [
    {'Group': 'State Median',
     'Science': state_median['pct_at_above_expectations_science'],
     'ELA': state_median['pct_at_above_expectations_ela'],
     'Math': state_median['pct_at_above_expectations_math']},
    {'Group': 'State Average',
     'Science': state_average['pct_at_above_expectations_science'],
     'ELA': state_average['pct_at_above_expectations_ela'],
     'Math': state_average['pct_at_above_expectations_math']},
    {'Group': 'South Portland Schools',
     'Science': sp_performance.get('pct_at_above_expectations_science'),
     'ELA': sp_performance.get('pct_at_above_expectations_ela'),
     'Math': sp_performance.get('pct_at_above_expectations_math')}
]

comparison_data
# Extract data for plotting
groups = [item['Group'] for item in comparison_data]
science_scores = [item['Science'] for item in comparison_data]
ela_scores = [item['ELA'] for item in comparison_data]
math_scores = [item['Math'] for item in comparison_data]

# Set up the bar chart
x = np.arange(len(groups))
width = 0.25

fig, ax = plt.subplots(figsize=(12, 7))

# Create bars
bars1 = ax.bar(x - width, science_scores, width, label='Science', color='#3498db', alpha=0.8)
bars2 = ax.bar(x, ela_scores, width, label='ELA', color='#2ecc71', alpha=0.8)
bars3 = ax.bar(x + width, math_scores, width, label='Math', color='#e74c3c', alpha=0.8)

# Add value labels on bars
def add_value_labels(bars):
    for bar in bars:
        height = bar.get_height()
        if height is not None and not np.isnan(height):
            ax.text(bar.get_x() + bar.get_width()/2., height,
                   f'{height:.1f}%',
                   ha='center', va='bottom', fontsize=9, fontweight='bold')

add_value_labels(bars1)
add_value_labels(bars2)
add_value_labels(bars3)

# Customize the chart
ax.set_xlabel('Group', fontsize=12, fontweight='bold')
ax.set_ylabel('Percentage at or Above Expectations', fontsize=12, fontweight='bold')
ax.set_title('2025 ESSA Results\nScience, ELA, and Math', 
             fontsize=14, fontweight='bold', pad=20)
ax.set_xticks(x)
ax.set_xticklabels(groups, fontsize=11)
ax.legend(fontsize=11, loc='upper left')
ax.set_ylim(0, max(max(science_scores), max(ela_scores), max(math_scores)) * 1.15)

# Add grid for easier reading
ax.grid(axis='y', alpha=0.3, linestyle='--')
ax.set_axisbelow(True)

plt.tight_layout()
plt.show()
```

## ESSA Performance: By Grade

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load the CSV file
df = pd.read_csv('essa_data/2025_maine_essa.csv')

# Define the latest year and key performance columns
latest_year = df['year'].max()
perf_cols = [
    'pct_at_above_expectations_science',
    'pct_at_above_expectations_ela',
    'pct_at_above_expectations_math'
]

# Get all unique grade levels (excluding 'All Grades')
grade_levels = df[df['grades'] != 'All Grades']['grades'].unique()
grade_levels = sorted([g for g in grade_levels if pd.notna(g)])

# Create a plot for each grade level
for grade in grade_levels:
    # Filter for the latest year, All Schools, All Students, specific grade
    district_grade_df = df[
        (df['year'] == latest_year) &
        (df['school'] == 'All Schools') &
        (df['students'] == 'All Students') &
        (df['grades'] == grade)
    ].copy()
    
    # Drop rows where any of the percentage columns is NaN
    district_grade_df.dropna(subset=perf_cols, inplace=True)
    
    # Check if we have data for this grade
    if len(district_grade_df) == 0:
        print(f"No data available for grade: {grade}")
        continue
    
    # 1. Calculate Statewide Performance for this grade
    state_average = district_grade_df[perf_cols].mean().to_dict()
    state_median = district_grade_df[perf_cols].median().to_dict()
    
    # 2. Extract South Portland Schools performance for this grade
    sp_df = district_grade_df[district_grade_df['district'] == 'South Portland Public Schools']
    
    if len(sp_df) == 0:
        print(f"No South Portland data for grade: {grade}")
        continue
    
    sp_performance = sp_df[perf_cols].iloc[0].to_dict()
    
    # 3. Combine the data
    comparison_data = [
        {'Group': 'State Median',
         'Science': state_median['pct_at_above_expectations_science'],
         'ELA': state_median['pct_at_above_expectations_ela'],
         'Math': state_median['pct_at_above_expectations_math']},
        {'Group': 'State Average',
         'Science': state_average['pct_at_above_expectations_science'],
         'ELA': state_average['pct_at_above_expectations_ela'],
         'Math': state_average['pct_at_above_expectations_math']},
        {'Group': 'South Portland Schools',
         'Science': sp_performance.get('pct_at_above_expectations_science'),
         'ELA': sp_performance.get('pct_at_above_expectations_ela'),
         'Math': sp_performance.get('pct_at_above_expectations_math')}
    ]
    
    # Extract data for plotting
    groups = [item['Group'] for item in comparison_data]
    science_scores = [item['Science'] for item in comparison_data]
    ela_scores = [item['ELA'] for item in comparison_data]
    math_scores = [item['Math'] for item in comparison_data]
    
    # Set up the bar chart
    x = np.arange(len(groups))
    width = 0.25
    fig, ax = plt.subplots(figsize=(12, 7))
    
    # Create bars
    bars1 = ax.bar(x - width, science_scores, width, label='Science', color='#3498db', alpha=0.8)
    bars2 = ax.bar(x, ela_scores, width, label='ELA', color='#2ecc71', alpha=0.8)
    bars3 = ax.bar(x + width, math_scores, width, label='Math', color='#e74c3c', alpha=0.8)
    
    # Add value labels on bars
    def add_value_labels(bars):
        for bar in bars:
            height = bar.get_height()
            if height is not None and not np.isnan(height):
                ax.text(bar.get_x() + bar.get_width()/2., height,
                       f'{height:.1f}%',
                       ha='center', va='bottom', fontsize=9, fontweight='bold')
    
    add_value_labels(bars1)
    add_value_labels(bars2)
    add_value_labels(bars3)
    
    # Customize the chart
    ax.set_xlabel('Group', fontsize=12, fontweight='bold')
    ax.set_ylabel('Percentage at or Above Expectations', fontsize=12, fontweight='bold')
    ax.set_title(f'2025 ESSA Results - Grade {grade}\nScience, ELA, and Math', 
                 fontsize=14, fontweight='bold', pad=20)
    ax.set_xticks(x)
    ax.set_xticklabels(groups, fontsize=11)
    ax.legend(fontsize=11, loc='upper left')
    
    # Calculate max value for consistent y-axis scaling
    max_val = max([s for s in science_scores + ela_scores + math_scores if s is not None and not np.isnan(s)])
    ax.set_ylim(0, max_val * 1.15)
    
    # Add grid for easier reading
    ax.grid(axis='y', alpha=0.3, linestyle='--')
    ax.set_axisbelow(True)
    
    plt.tight_layout()
    plt.show()
```

## Compared to Neighboring School Districts

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load the CSV file
df = pd.read_csv('essa_data/2025_maine_essa.csv')

# Define target districts for comparison
target_districts = [
    'South Portland Public Schools',
    'Yarmouth Public Schools',
    'Falmouth Public Schools',
    'Portland Public Schools',
    'Westbrook Public Schools',
    'Scarborough Public Schools',
    'Cape Elizabeth Public Schools'
]

# Define the latest year and key performance columns
latest_year = df['year'].max()
perf_cols = [
    'pct_at_above_expectations_science',
    'pct_at_above_expectations_ela',
    'pct_at_above_expectations_math'
]

# Filter for the latest year, All Schools, All Students, All Grades
district_aggregate_df = df[
    (df['year'] == latest_year) &
    (df['school'] == 'All Schools') &
    (df['students'] == 'All Students') &
    (df['grades'] == 'All Grades') &
    (df['district'].isin(target_districts))
].copy()

# Drop rows where any of the percentage columns is NaN
district_aggregate_df.dropna(subset=perf_cols, inplace=True)

# Sort alphabetically for consistent display
district_aggregate_df = district_aggregate_df.sort_values('district')

# Extract data for plotting
districts = district_aggregate_df['district'].tolist()
science_scores = district_aggregate_df['pct_at_above_expectations_science'].tolist()
ela_scores = district_aggregate_df['pct_at_above_expectations_ela'].tolist()
math_scores = district_aggregate_df['pct_at_above_expectations_math'].tolist()

# Create shortened labels (remove "Public Schools" for readability)
short_labels = [d.replace(' Public Schools', '') for d in districts]

# Set up the bar chart
x = np.arange(len(districts))
width = 0.25

fig, ax = plt.subplots(figsize=(14, 8))

# Create bars
bars1 = ax.bar(x - width, science_scores, width, label='Science', color='#3498db', alpha=0.8)
bars2 = ax.bar(x, ela_scores, width, label='ELA', color='#2ecc71', alpha=0.8)
bars3 = ax.bar(x + width, math_scores, width, label='Math', color='#e74c3c', alpha=0.8)

# Add value labels on bars
def add_value_labels(bars):
    for bar in bars:
        height = bar.get_height()
        if height is not None and not np.isnan(height):
            ax.text(bar.get_x() + bar.get_width()/2., height,
                   f'{height:.1f}%',
                   ha='center', va='bottom', fontsize=9, fontweight='bold')

add_value_labels(bars1)
add_value_labels(bars2)
add_value_labels(bars3)

# Highlight South Portland with a subtle background
try:
    south_portland_idx = districts.index('South Portland Public Schools')
    ax.axvspan(south_portland_idx - 0.5, south_portland_idx + 0.5, 
               alpha=0.1, color='gold', zorder=0)
except ValueError:
    pass  # South Portland not in list

# Customize the chart
ax.set_xlabel('District', fontsize=12, fontweight='bold')
ax.set_ylabel('Percentage at or Above Expectations', fontsize=12, fontweight='bold')
ax.set_title('2025 ESSA Results: South Portland vs Neighboring Districts\nScience, English Language Arts, and Math', 
             fontsize=14, fontweight='bold', pad=20)
ax.set_xticks(x)
ax.set_xticklabels(short_labels, fontsize=10, rotation=45, ha='right')
ax.legend(fontsize=11, loc='upper left')

# Set y-axis limit with some headroom
all_scores = science_scores + ela_scores + math_scores
ax.set_ylim(0, max(all_scores) * 1.15)

# Add grid for easier reading
ax.grid(axis='y', alpha=0.3, linestyle='--')
ax.set_axisbelow(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load the CSV file
df = pd.read_csv('essa_data/2025_maine_essa.csv')

# Define target districts for comparison
target_districts = ['Lewiston Public Schools', 'Biddeford Public Schools', 'Sanford Public Schools', 'Waterville Public Schools', 'South Portland Public Schools']


# Define the latest year and key performance columns
latest_year = df['year'].max()
perf_cols = [
    'pct_at_above_expectations_science',
    'pct_at_above_expectations_ela',
    'pct_at_above_expectations_math'
]

# Filter for the latest year, All Schools, All Students, All Grades
district_aggregate_df = df[
    (df['year'] == latest_year) &
    (df['school'] == 'All Schools') &
    (df['students'] == 'All Students') &
    (df['grades'] == 'All Grades') &
    (df['district'].isin(target_districts))
].copy()

# Drop rows where any of the percentage columns is NaN
district_aggregate_df.dropna(subset=perf_cols, inplace=True)

# Sort alphabetically for consistent display
district_aggregate_df = district_aggregate_df.sort_values('district')

# Extract data for plotting
districts = district_aggregate_df['district'].tolist()
science_scores = district_aggregate_df['pct_at_above_expectations_science'].tolist()
ela_scores = district_aggregate_df['pct_at_above_expectations_ela'].tolist()
math_scores = district_aggregate_df['pct_at_above_expectations_math'].tolist()

# Create shortened labels (remove "Public Schools" for readability)
short_labels = [d.replace(' Public Schools', '') for d in districts]

# Set up the bar chart
x = np.arange(len(districts))
width = 0.25

fig, ax = plt.subplots(figsize=(14, 8))

# Create bars
bars1 = ax.bar(x - width, science_scores, width, label='Science', color='#3498db', alpha=0.8)
bars2 = ax.bar(x, ela_scores, width, label='ELA', color='#2ecc71', alpha=0.8)
bars3 = ax.bar(x + width, math_scores, width, label='Math', color='#e74c3c', alpha=0.8)

# Add value labels on bars
def add_value_labels(bars):
    for bar in bars:
        height = bar.get_height()
        if height is not None and not np.isnan(height):
            ax.text(bar.get_x() + bar.get_width()/2., height,
                   f'{height:.1f}%',
                   ha='center', va='bottom', fontsize=9, fontweight='bold')

add_value_labels(bars1)
add_value_labels(bars2)
add_value_labels(bars3)

# Highlight South Portland with a subtle background
try:
    south_portland_idx = districts.index('South Portland Public Schools')
    ax.axvspan(south_portland_idx - 0.5, south_portland_idx + 0.5, 
               alpha=0.1, color='gold', zorder=0)
except ValueError:
    pass  # South Portland not in list

# Customize the chart
ax.set_xlabel('District', fontsize=12, fontweight='bold')
ax.set_ylabel('Percentage at or Above Expectations', fontsize=12, fontweight='bold')
ax.set_title('2025 ESSA Results: South Portland vs Neighboring Districts\nScience, English Language Arts, and Math', 
             fontsize=14, fontweight='bold', pad=20)
ax.set_xticks(x)
ax.set_xticklabels(short_labels, fontsize=10, rotation=45, ha='right')
ax.legend(fontsize=11, loc='upper left')

# Set y-axis limit with some headroom
all_scores = science_scores + ela_scores + math_scores
ax.set_ylim(0, max(all_scores) * 1.15)

# Add grid for easier reading
ax.grid(axis='y', alpha=0.3, linestyle='--')
ax.set_axisbelow(True)

plt.tight_layout()
plt.show()
```
